# excel to scv

```python
import os
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/content/drive/MyDrive/valid-might-460212-k2-aabca356a049.json'

import tempfile
from google.cloud import storage
import pandas as pd

GCS_BUCKET = 'de-project2-bucket-1'
GCS_XLSX_PREFIX = 'kyobo/excel'
GCS_CSV_PREFIX = 'kyobo/csv'

client = storage.Client()
bucket = client.bucket(GCS_BUCKET)

def get_standardized_filename(filename):
    # meta_010104_20250519_065002.xlsx → meta_010104.xlsx
    # meta_05132127.xlsx → meta_05132127.xlsx
    base = os.path.basename(filename)
    if base.count('_') >= 2:
        # meta_010104_20250519_065002.xlsx → meta_010104.xlsx
        parts = base.split('_')
        new_base = '_'.join(parts[:2]) + '.xlsx'
        return new_base
    else:
        return base

with tempfile.TemporaryDirectory() as tmpdir:
    blobs = client.list_blobs(GCS_BUCKET, prefix=GCS_XLSX_PREFIX)
    for blob in blobs:
        if not blob.name.endswith('.xlsx'):
            continue
        print(f"Processing: {blob.name}")
        std_xlsx_name = get_standardized_filename(blob.name)
        local_xlsx = os.path.join(tmpdir, std_xlsx_name)
        local_csv = os.path.join(tmpdir, os.path.splitext(std_xlsx_name)[0] + '.csv')
        # 2. 다운로드
        blob.download_to_filename(local_xlsx)
        # 3. 엑셀 → CSV 변환
        try:
            df = pd.read_excel(local_xlsx)
            df.columns = [
                'index', 'product_code', 'product_id', 'title', 'list_price',
                'sale_price', 'discount_rate', 'point_rate', 'expected_points',
                'author', 'publisher', 'pub_date', 'category'
            ]
            df.to_csv(local_csv, index=False, encoding='utf-8-sig')
        except Exception as e:
            print(f"Failed to convert {blob.name}: {e}")
            continue
        # 4. CSV를 GCS에 업로드
        csv_blob_name = f"{GCS_CSV_PREFIX}/{os.path.splitext(std_xlsx_name)[0]}.csv"
        bucket.blob(csv_blob_name).upload_from_filename(local_csv)
        print(f"Uploaded CSV: {csv_blob_name}")
```

# merge image_url and meta data by "product_id"

```python
import os
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/content/drive/MyDrive/valid-might-460212-k2-aabca356a049.json'

import tempfile
from google.cloud import storage
import pandas as pd

GCS_BUCKET = 'de-project2-bucket-1'
GCS_XLSX_PREFIX = 'kyobo/excel'
GCS_CSV_PREFIX = 'kyobo/csv'

client = storage.Client()
bucket = client.bucket(GCS_BUCKET)
```

```python
from google.cloud import storage
import pandas as pd
import tempfile

client = storage.Client()
bucket = client.bucket(GCS_BUCKET)

# 1. 모든 CSV 파일 통합
# 모든 CSV 파일을 담을 빈 DataFrame 생성
combined_df = pd.DataFrame()

# GCS에서 CSV 파일 목록 가져오기
csv_blobs = client.list_blobs(GCS_BUCKET, prefix=GCS_CSV_PREFIX)

for blob in csv_blobs:
    if not blob.name.endswith('.csv'):
        continue
    print(f"Loading: {blob.name}")

    # 임시 파일로 다운로드
    with tempfile.NamedTemporaryFile(delete=False) as temp_file:
        blob.download_to_filename(temp_file.name)

        # CSV 읽기
        df = pd.read_csv(temp_file.name, encoding='utf-8-sig')
        combined_df = pd.concat([combined_df, df], ignore_index=True)

print("총 데이터 개수:", len(combined_df))
```

```python
# 2. 실제 GCS 이미지 URL 매핑
# --------------------------------------------------
# 이미지 파일 리스트 조회
image_dict = {}
PREFIX = "kyobo/images/"
image_blobs = bucket.list_blobs(prefix=PREFIX)

for blob in image_blobs:
    if blob.name.endswith('.jpg'):
        # 경로에서 product_id 추출 (예: kyobo/images/010101/S000000408032.jpg)
        parts = blob.name.split('/')
        if len(parts) >= 4:
            product_id = parts[-1].replace('.jpg', '')
            image_url = f"https://storage.googleapis.com/{GCS_BUCKET}/{blob.name}"
            image_dict[product_id] = image_url

# 3. 데이터 타입 변환 및 매핑
combined_df['product_id'] = combined_df['product_id'].astype(str)
combined_df['image_url'] = combined_df['product_id'].map(image_dict)

# 4. 결과 확인
print("\n🔍 최종 데이터 샘플:")
print(combined_df[['product_id', 'image_url']].head(3))

# 5. 통합 CSV 저장 및 업로드
with tempfile.TemporaryDirectory() as tmpdir:
    local_combined = os.path.join(tmpdir, "combined_data.csv")
    combined_df.to_csv(local_combined, index=False, encoding='utf-8-sig')

    # GCS 업로드
    upload_blob = bucket.blob(f"{GCS_CSV_PREFIX}/combined/combined_data.csv")
    upload_blob.upload_from_filename(local_combined)
    print("\n✅ 통합 CSV 업로드 완료!")
    print(f"경로: gs://{GCS_BUCKET}/{GCS_CSV_PREFIX}/combined/combined_data.csv")
```

```python
# image_url 컬럼에서 NaN과 빈 문자열('')을 제외한 값의 개수 세기
image_url_count = combined_df['image_url'].dropna().apply(lambda x: x != '').sum()
print(f"이미지 URL이 존재하는 행 개수: {image_url_count}")
```

```python
# image_url이 NaN이거나 빈 문자열('')인 행만 추출
no_image_df = combined_df[combined_df['image_url'].isna() | (combined_df['image_url'] == '')]

# 결과 확인
print(no_image_df.shape)
```

```python
# no_image_df.to_csv('no_image_rows.csv', index=False)

# GCS
import tempfile

# 1. 로컬(Colab) 임시 파일로 저장
with tempfile.NamedTemporaryFile(suffix='.csv', delete=False) as temp_file:
    no_image_df.to_csv(temp_file.name, index=False, encoding='utf-8-sig')
    local_csv_path = temp_file.name

# 2. GCS에 업로드
output_blob = bucket.blob(f"{GCS_CSV_PREFIX}/combined/no_image.csv")
output_blob.upload_from_filename(local_csv_path)
print("✅ 전처리 후 CSV 업로드 완료!")
print(f"경로: gs://{GCS_BUCKET}/{GCS_CSV_PREFIX}/combined/no_image.csv")
```

```python
# 1. product_id 기준 중복된 행의 개수(첫 번째만 남기고 나머지 카운트)
dup_count = combined_df.duplicated(subset=['product_id']).sum()
print(f"product_id 기준 중복된 행 개수: {dup_count}")

# 2. product_id별로 몇 번씩 나오는지 보고 싶다면
dup_detail = combined_df['product_id'].value_counts()
dup_detail = dup_detail[dup_detail > 1]
print(f"\n2번 이상 등장하는 product_id 개수: {len(dup_detail)}")
print(dup_detail.head())
```

```python
# 다시 전처리 진행
# 1. image_url이 없는 행 제거
# image_url이 NaN이거나 빈 문자열('')인 행 제거
filtered_df = combined_df.dropna(subset=['image_url'])
filtered_df = filtered_df[filtered_df['image_url'] != '']

# 2. product_id 중복 제거 (하나만 남김)
filtered_df = filtered_df.drop_duplicates(subset=['product_id'], keep='first')

print(f"정제 후 데이터 개수: {len(filtered_df)}")
print(filtered_df.head())
```

# excel to scv

```python
import os
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/content/drive/MyDrive/valid-might-460212-k2-aabca356a049.json'

import tempfile
from google.cloud import storage
import pandas as pd

GCS_BUCKET = 'de-project2-bucket-1'
GCS_XLSX_PREFIX = 'kyobo/excel'
GCS_CSV_PREFIX = 'kyobo/csv'

client = storage.Client()
bucket = client.bucket(GCS_BUCKET)

def get_standardized_filename(filename):
    # meta_010104_20250519_065002.xlsx  meta_010104.xlsx
    # meta_05132127.xlsx  meta_05132127.xlsx
    base = os.path.basename(filename)
    if base.count('_') >= 2:
        # meta_010104_20250519_065002.xlsx  meta_010104.xlsx
        parts = base.split('_')
        new_base = '_'.join(parts[:2]) + '.xlsx'
        return new_base
    else:
        return base

with tempfile.TemporaryDirectory() as tmpdir:
    blobs = client.list_blobs(GCS_BUCKET, prefix=GCS_XLSX_PREFIX)
    for blob in blobs:
        if not blob.name.endswith('.xlsx'):
            continue
        print(f"Processing: {blob.name}")
        std_xlsx_name = get_standardized_filename(blob.name)
        local_xlsx = os.path.join(tmpdir, std_xlsx_name)
        local_csv = os.path.join(tmpdir, os.path.splitext(std_xlsx_name)[0] + '.csv')
        # 2. 
        blob.download_to_filename(local_xlsx)
        # 3.   CSV 
        try:
            df = pd.read_excel(local_xlsx)
            df.columns = [
                'index', 'product_code', 'product_id', 'title', 'list_price',
                'sale_price', 'discount_rate', 'point_rate', 'expected_points',
                'author', 'publisher', 'pub_date', 'category'
            ]
            df.to_csv(local_csv, index=False, encoding='utf-8-sig')
        except Exception as e:
            print(f"Failed to convert {blob.name}: {e}")
            continue
        # 4. CSV GCS 
        csv_blob_name = f"{GCS_CSV_PREFIX}/{os.path.splitext(std_xlsx_name)[0]}.csv"
        bucket.blob(csv_blob_name).upload_from_filename(local_csv)
        print(f"Uploaded CSV: {csv_blob_name}")
```

```python
import os
os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = '/content/drive/MyDrive/valid-might-460212-k2-aabca356a049.json'

import tempfile
from google.cloud import storage
import pandas as pd

GCS_BUCKET = 'de-project2-bucket-1'
GCS_XLSX_PREFIX = 'kyobo/excel'
GCS_CSV_PREFIX = 'kyobo/csv'

client = storage.Client()
bucket = client.bucket(GCS_BUCKET)
```

```python
from google.cloud import storage
import pandas as pd
import tempfile

client = storage.Client()
bucket = client.bucket(GCS_BUCKET)

# 1.  CSV  
#  CSV    DataFrame 
combined_df = pd.DataFrame()

# GCS CSV   
csv_blobs = client.list_blobs(GCS_BUCKET, prefix=GCS_CSV_PREFIX)

for blob in csv_blobs:
    if not blob.name.endswith('.csv'):
        continue
    print(f"Loading: {blob.name}")

    #   
    with tempfile.NamedTemporaryFile(delete=False) as temp_file:
        blob.download_to_filename(temp_file.name)

        # CSV 
        df = pd.read_csv(temp_file.name, encoding='utf-8-sig')
        combined_df = pd.concat([combined_df, df], ignore_index=True)

print("  :", len(combined_df))
```

```python
# 2.  GCS  URL 
# --------------------------------------------------
#    
image_dict = {}
PREFIX = "kyobo/images/"
image_blobs = bucket.list_blobs(prefix=PREFIX)

for blob in image_blobs:
    if blob.name.endswith('.jpg'):
        #  product_id  (: kyobo/images/010101/S000000408032.jpg)
        parts = blob.name.split('/')
        if len(parts) >= 4:
            product_id = parts[-1].replace('.jpg', '')
            image_url = f"https://storage.googleapis.com/{GCS_BUCKET}/{blob.name}"
            image_dict[product_id] = image_url

# 3.     
combined_df['product_id'] = combined_df['product_id'].astype(str)
combined_df['image_url'] = combined_df['product_id'].map(image_dict)

# 4.  
print("\n   :")
print(combined_df[['product_id', 'image_url']].head(3))

# 5.  CSV   
with tempfile.TemporaryDirectory() as tmpdir:
    local_combined = os.path.join(tmpdir, "combined_data.csv")
    combined_df.to_csv(local_combined, index=False, encoding='utf-8-sig')

    # GCS 
    upload_blob = bucket.blob(f"{GCS_CSV_PREFIX}/combined/combined_data.csv")
    upload_blob.upload_from_filename(local_combined)
    print("\n  CSV  !")
    print(f": gs://{GCS_BUCKET}/{GCS_CSV_PREFIX}/combined/combined_data.csv")
```
